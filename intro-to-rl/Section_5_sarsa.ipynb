## Import the necessary software libraries:

```python
import numpy as np
import matplotlib.pyplot as plt

from envs import Maze
from utils import plot_policy, plot_action_values, test_agent
```

## Create the environment, value table and policy

#### Create the environment

```python
env = Maze()
```

#### Create the $Q(s, a)$ table

```python
action_values = np.zeros((5,5,4))
```

#### Create the policy $\pi(s)$

```python
def policy(state, epsilon=0.2):
    if np.random.random() < epsilon:
        return np.random.randint(4)
    av = action_values[state]
    return np.random.choice(np.flatnonzero(av == av.max()))
```

#### Plot the value table $Q(s,a)$

```python
plot_action_values(action_values)
```

#### Plot the policy

```python
plot_policy(action_values, env.render(mode='rgb_array'))
```

## Implement the algorithm

</br>



<div style="text-align:center">
    Adapted from Barto & Sutton: "Reinforcement Learning: An Introduction".
</div>

```python
def sarsa(action_values, policy, episodes, alpha=0.1, gamma=0.99, epsilon=0.2):
    for episode in range(1, episodes + 1):
        state = env.reset()
        action = policy(state, epsilon)
        done = False
        
        while not done:
            next_state, reward, done, _ = env.step(action)
            next_action = policy(next_state, epsilon)
            
            next_qsa = action_values[next_state][next_action]
            current_qsa = action_values[state][action]
            action_values[state][action] = current_qsa + alpha * (reward + gamma * next_qsa - current_qsa)
            
            state = next_state
            action = next_action
```

```python
sarsa(action_values, policy, episodes=10000)
```

## Show results

#### Show resulting value table $Q(s,a)$

```python
plot_action_values(action_values)
```

#### Show resulting policy $\pi(\cdot|s)$

```python
plot_policy(action_values, env.render(mode='rgb_array'))
```

#### Test the resulting agent

```python
test_agent(env, policy)
```
